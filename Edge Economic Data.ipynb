```python
#The following is a group of dataframes cleaned by Juls Marino for Team-82.

df1 = pd.read_csv('EDGE_ACS_CWIFT2018_State.txt', sep = '\t', header=None) #This is the data by State.

#The CWIFT measures wage and salary differences for college graduates. 
#The CWIFT for each location isthe predicted wage level for that location divided by the
# national average predicted wage,
df1.columns = ['ST_FIPS','ST_NAME','ST_CWIFTEST', 'ST_CWIFTSE']
#So we can call certain columns by name.

df1['ST_CWIFTEST'] = pd.to_numeric(df1['ST_CWIFTEST'],errors='coerce')
df1['ST_CWIFTSE'] = pd.to_numeric(df1['ST_CWIFTSE'],errors='coerce')
#This will help us use these values as floats, not string objects.

df1 = df1.rename(columns={'ST_NAME':'State', 'ST_CWIFTEST':'CWIFTEST_Score', 
                          'ST_CWIFTSE':'CWIFTEST_Standard_Error'})


df1.drop(df1.index[0], axis=0)
#This is meant to drop the first row so the column names aren't duplicated,
#but it isn't working yet. Hence the 'NaN' in the first row.

df1.drop(columns='ST_FIPS', inplace=True) 
#This is the two-digit Federal Information Processing Standard (FIPS)
#code for the state, which we don't need.

df1 = df1.set_index('State')
#This will make it easier to index by state name.


df1
```
